```python
from __future__ import division

import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

n_train = 5400  # samples for training
n_test = 600  # samples for testing
n_features = 1326
n_features_max = 75  # maximum number of features
step = 4  # step size for the calculation


def generate_data(n_samples, n_features):
    """Generate random blob-ish data with noisy features.

    This returns an array of input data with shape `(n_samples, n_features)`
    and an array of `n_samples` target labels.

    Only one feature contains discriminative information, the other features
    contain only noise.
    """
    X, y = make_blobs(n_samples=n_samples, n_features=1, centers=[[-0.5], [0.5]])

    # add non-discriminative features
    if n_features > 1:
        X = np.hstack([X, np.random.randn(n_samples, n_features - 1)])
    return X, y

X_train, y_train = generate_data(n_train, n_features)

# clf1 = LinearDiscriminantAnalysis( ).fit(X_train, y_train)
clf1 = LinearDiscriminantAnalysis(solver='lsqr', shrinkage=None).fit(X_train, y_train)

X, y = generate_data(n_test, n_features)
score_clf1 = clf1.score(X, y)

plt.subplot(211)
plt.scatter(X_train[:,0],X_train[:,1],c=y_train)
plt.subplot(212)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()

print('score is', score_clf1)
```
